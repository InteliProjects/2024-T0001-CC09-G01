## Distribuição de rentabilidades ideal

Para gerar o gráfico foi utilizada a distribuição Beta. Ela é definida no intervalo [0,1] denotado por α e β, geralmente. α e β são dois parâmetros positivos que aparecem como expoentes da variável aleatória e têm o objetivo de controlar a forma da distribuição. Sua notação é Beta(α,β), em que α e β são números reais, e os valores são maiores que zero. O uso mais comum dessa distribuição é para modelar a incerteza sobre a probabilidade de sucesso de um experimento aleatório. Contudo, ela foi escolhida pois seus parâmetros permitem alterar a forma mais precisamente, aproximando-se melhor da distribuição esperada definida a partir da modelagem e do que foi discutido com o parceiro do projeto.

A fórmula para a função de densidade de probabilidade é:

$$ f(x) = x^{\alpha - 1}(1-x)^{\beta-1} / B(\alpha, \beta)$$
$$ B(\alpha, \beta) = \int_0^1 t^{\alpha-1}(1-t)^{\beta-1}dt$$

A curva laranja representa a curva mais ideal, em que o pico está bem próximo de 100% do CDI e ela decresce rapidamente para os dois lados. Contudo, na realidade será muito difícil acontecer isso. Por isso, a curva azul representa uma distribuição mais realista, em conformidade com a modelagem e as funções de recompensa do grupo. Neste caso, a curva tem o pico entre 100% e 105% (recompensa máxima), e decresce proporcionalmente até 98% e 120% (recompensa proporcional). Valores abaixo desse intervalo são penalizados. 

```python
# Define the interval
a, b = 98, 120

# Define shape parameters
# Start with alpha = beta for simplicity; adjust as needed to fit the peak and spread
alpha, beta_value = 1.5, 3.5  # These are just example values; tweak these based on your needs

# Generate x values from a to b, which will be used to compute the PDF values
x = np.linspace(a, b, 1000)

# Compute the PDF values for each x
# The beta distribution is defined on [0, 1], so we need to scale and shift x values accordingly
y_1 = beta.pdf((x - a) / (b - a), alpha, beta_value) / (b - a)

a = 99
x = np.linspace(a, b, 1000)
alpha, beta_value = 1.3, 7
y_2 = beta.pdf((x - a) / (b - a), alpha, beta_value) / (b - a)

# Plotting the PDF
plt.figure(figsize=(10, 6))
plt.plot(x, y_1, lw=2, label="grupo")
plt.plot(x, y_2, lw=2)
plt.title('Distribuições Ideais ')
plt.xlabel('Rentabilidade (%)')
plt.ylabel('Densidade')
plt.legend(["intervalo de recompensas da modelagem", "foco máximo no 100% do CDI"])
plt.grid()
plt.show()
```

## Distribuição de variância ideal

Outra distribuição é a de variância dos preços das compras de uma combinação. O intuito é mostrar que uma maior variância será priorizada para gerar uma maior diversificação de compras em uma combinação.

OBS: o intervalo do eixo x pode ser alterado posteriormente para se adequar aos dados reais.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# Define o intervalo
a, b = 0, 0.5

# Define os parâmetros da distribuição beta para ajuste
alpha, beta_value = 3, 4.5  # Ajuste conforme necessário para encontrar a distribuição ideal

# Gera os valores de x do intervalo
x = np.linspace(a, b, 1000)

# Calcula os valores da função de densidade de probabilidade (PDF) para cada x
# A distribuição beta é definida em [0, 1], então precisamos escalonar e deslocar os valores de x adequadamente
y = beta.pdf((x - a) / (b - a), alpha, beta_value) / (b - a)

# Plota a PDF
plt.figure(figsize=(10, 6))
plt.plot(x, y, lw=2)
plt.title('Distribuição de Variância Ajustada')
plt.xlabel('Variância')
plt.ylabel('Densidade')
plt.grid(axis='y')
plt.xlim([a, b])
plt.show()
```
